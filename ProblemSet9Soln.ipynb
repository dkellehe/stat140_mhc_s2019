(a) Write the equation of the regression line for predicting travel time. 

```r
113*(0.636)/99
```

Answer: $b_1 = \frac{113}{99}\times 0.636 \approx 0.726$. There also seems to be a point that is about $(110,120)$, so we approximate the line with
$$ y-120 = 0.726(x-110)$$
