## Problem 1 - Using summary of a linear regression

Here I've loaded in the data from the iris data and created `iris.setosa` the iris data set which only has the setosa species (so we don't have to worry about Simpson's paradox)

```r
#Run but do not change
require(oibiostat)
require(ggplot2)
require(dplyr)

#Here is the data that we will be looking at
data(iris)
iris.setosa <- filter(iris,Species=='setosa')
ggplot(iris.setosa,aes(x=Sepal.Length,y=Sepal.Width))+geom_point()
```

a) Create a linear model from `iris.setosa` with x= `Sepal.Length` (predictor variable), and y= `Sepal.Width` (target variable) and print it's `summary`. 

```r
# Calculation here
sepal.shape = lm(Sepal.Width~Sepal.Length, data=iris.setosa)
summary(sepal.shape)
```

b) What is the slope and y-intercept of the linear model? Interpret the slope in terms of the data.

Answer: Slope is 0.80 and intercepts is -0.57. For every cm longer that the sepal is, it is 4/5 of a cm wider.

c) What is the probability that we could have seen at least as strong of an association by chance? (i.e. find the $p$-value)

Answer: $6.71\times10^{-10}$.

d) What is the IQR of the residuals?

Answer:  $0.15738 -(-0.18273)\approx 0.34$

e) Is the $R^2$ high enough to make this a good fit? Explain.

Answer: Yes, 0.55>0.5.
