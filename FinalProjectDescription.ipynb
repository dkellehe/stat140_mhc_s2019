Congratulations, you have made it to the end of the semester, and this is the final project.

The project will consist of a short 10 minutes presentation and a jupyter notebook documenting the code that you used.

## Project goals

In contrast with the midterm project, the goal for the final project is to use the article as a starting point, and to create a presentation which is based on something that interested you about the data or the subject of the article.

## Content of the presentation

Your presentation should 

1) Summarize the article

- What are the hypotheses and results?

- How are graphs/data/visualisations used to support these?

2) What questions did you have after reading the article?

3) How did you use statistics and data to investigate these questions?

- Your investigation should have 2 visualisations, at least one should be advanced (i.e. involve regression or confidence intervals) and at least one should come from a secondary/outside datasource.
